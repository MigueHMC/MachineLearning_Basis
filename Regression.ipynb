```python
import pandas as pd

import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('C:/Users/migue/OneDrive/Documentos/MachineLearning_Basis/historical_prem_standings.csv')
```

```python
df.head()
```

```python
X = df['GF'].values.reshape(-1,1)
y = df['Rk'].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
plt.scatter(X, y, color='blue')
plt.plot(X_test, y_pred, color='red')
plt.xlabel('Goals For (GF)')
plt.ylabel('Final Position (Rk)')
plt.title('Goals For vs Position')
plt.ylim(0, 21)
plt.show()
```

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
```

```python
print('R-squared:', model.score(X, y))
```

```python
predictions = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```
