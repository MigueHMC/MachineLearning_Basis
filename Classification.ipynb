```python
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('C:/Users/migue/OneDrive/Documentos/MachineLearning_Basis/classification_data.csv')
```

```python
df.head()
```

```python
df['Pos'] = df['Pos'].apply(lambda x: x.split(',')[0])
```

```python
df['Pos'].unique()
```

```python
df = df[df['Pos'] != 'Pos']
```

```python
df.isnull().sum()
```

```python
df = df.dropna()

print(df.shape)
```

```python
df['Pos'].value_counts()
```

```python
df['Pos'].value_counts().max() / df['Pos'].value_counts().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()

df['Pos'] = le.fit_transform(df['Pos'])

print(df['Pos'].unique())
```

```python
print(le.inverse_transform(df['Pos'].unique()))
```

```python
df[df['Pos'] == 2]
```

```python
df.columns
```

```python
X = df[['Gls', 'Ast', 'G+A', 'G-PK', 'PK', 'PKatt',
       'CrdY', 'CrdR', 'xG', 'npxG', 'xAG', 'npxG+xAG', 'PrgC', 'PrgP',
       'PrgR']]

y = df['Pos']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier

clf = DecisionTreeClassifier()

clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = clf.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
# Confusion Matrix
from sklearn.metrics import confusion_matrix

confusion_matrix(y_test, y_pred)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```
